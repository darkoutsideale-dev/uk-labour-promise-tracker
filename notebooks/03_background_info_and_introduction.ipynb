# 1. Introduction and Motivation

### The Broad Idea: Scaling Democratic Accountability

In a modern democracy, politicians make various promises during elections, and citizens are voting for a candidate based on those promises. However, when the election is over, it is very difficult for everyday people to keep track of whether or not the government is fulfilling those promises. This is the cause for a significant gap in trust and information for people. 

Currently, verifying political promises made by politicians is done mostly by professional fact-checking organizations, such as Full Fact or PolitiFact. While these groups have been successful with their work, their process is almost fully manual, therefore, it requires journalists to spend hours going through news reports, finding documents, and writing summaries. This way, traditional promise tracking made by time consuming and expensive human labour, becomes difficult to scale.

Our project started with a broad idea: How can we use automation to make political tracking easier and faster? We wanted to explore if we could use public data sources, APIs, and code to turn this manual journalistic task into a more automated information system. 


### Our Focus on UK Labour (2024) and Housing

The UK Labour Party won a significant election victory in July of 20024. by making promises to reform the country. We specifically chose to focus on housing for our testings for two main reasons. The first one is that it should have a clear paper trail in Parliament, which would make it easy for us to follow. This would mean that we can test our code against structured public databases, like the UK Parliament Bills API and `legislation.gov.uk`. 

The second reason is that housing related information is backed by concrete numbers, which makes it highly measurable because the actual numbers (e.g. number of houses being built) eventually do show up in reports from the Office for National Statistics (ONS).

# 2. Background: Political Promises as Data Systems

### The Information Gap: Words vs. Databases

While looking at this project with an information science lens, we can see that the biggest issue with politicla accountability is that it is a **data mismatch problem**.

When a political party releases a manifesto, they are giving us **unstructured text**. This means that it is written in long and messy sentences meant for the human voters. Usually, it is full of political initiatives and complex language that the computer has trouble reading and sorting it. 

However, when the government does fulfill one of their promises, the evidence for it is stored as **structured data**. This data is hidden away inside of the official code databses (APIs) or spreadsheets from the Office for National Statistics (`ONS`).

The "information gap" appears because these two different sides do not interconnect. At the moment, everyday citizens need to manually read the news in order to connect the dots. The goal of our project is to build a bridge that connects these two sides, this way turning political words into a clear, trackable data channel.


### The Digital Paper Trail of The Manifesto 

We can fully follow one promise as it moves through our channel in order to understand it better. 

We will choose a random promise, for example Labour's promise `H11`: *"Abolish Section 21 no-fault evictions."*. This is what the promise will go through while being in our data system:

1. **Text Cleaning:** The system takes away the extra words and brings out the key search terms: `Section 21` and `no-fault evictions`.
2. **API Search:** The code takes the keywords and automatically searches the `UK Parliament Bills API`, looking for any new laws matching those terms.
3. **Reading the Code:** When it finds a match (like the *Renters' Rights Act 2025*), our system reads the raw data from the API to check its progress. 
4. **The Final Check:** The channel looks for a specific tag called **Royal Assent**. The Royal Assent is like a `True/False` switch. If it is `True`, it means the bill has successfully become an active law.

We can focus on this digital paper trail, and then we can test where automation works perfectly and where it gets stuck. The background of this project is not just about housing laws, but it is about testing how well code can connect unfounded political promises to real facts that can be read by machines.

# 3. Research Questions

### Research Question

In order to guide our project, we formulated a data-driven research question:

> **To what extent can UK Labour’s 2024 manifesto promises on housing be tracked through parliamentary, governmental and budget documents and policies?**

This question does not ask if the government is succeeding politically, but it tests the limits of our data collection methods to see if political accountability can be tracked using public data channels.


### Testable Sub-Questions

In order to answer our main question, we broke it down into two specific, testable sub-questions. These sub-questions match the experiments we ran in our Python code:

#### Sub-Question 1 (Data Matching & Feasibility)
> *How effectively can unstructured campaign text automatically be designed for structured databases using keyword-matching algorithms?*

This approach is completely measurable. In our code, we tested this by running a script that matched 114 keywords against 18 core housing promises. The data showed a **coverage rate of 55.6%**, successfully linking 10 out of 18 promises to active legislative documents.

#### Sub-Question 2 (Channel Reliability & Drawbacks)
> *Where do automated data retrieval channels fail due to technical API issues, and how can human curated data be used as a drawback in order to verify the tracking accuracy?*

This directly tests the methodology of data tracking. When we were exploring more with the research, our automated script faced a real-world `ReadTimeout` connection error with the live Parliament API. We tested a drawback solution by creating a human-curated dataset of known legislation (such as the *Renters' Rights Act 2025*) to get over the technical issue and complete the data matching.

Our pilot notebook was able to successfully load the primary dataset of 18 core housing promises and ran a keyword extraction script, generating 114 specific search terms. When we matched it against current UK legislative records, the data showed us some insights:

* **Initial Coverage:** The system achieved an initial **55.6% coverage rate**, automatically linking 10 out of the 18 manifesto promises to specific legal evidence.
* **Key Laws Passed:** The channel successfully linked campaign promises to parts of the legislation, such as matching tenant protections (`H11`, `H12`) to the *Renters' Rights Act 2025* and planning updates (`H02`, `H03`, `H05`, `H06`, `H17`, `H18`) to the *Planning and Infrastructure Act 2025*.
* **Identifying the Data Gaps:** For the 8 promises that returned "No Evidence" in the legal phase (e.g. the target to build 1.5 million homes), our pilot proved exactly why a complex chanel is necessary. These promises are not missing because the government forgot them, but because they rely more on budget allocations, rather than new laws. This shows where we need to continue the work to our next step, which is tracking budgets and economic numbers.
